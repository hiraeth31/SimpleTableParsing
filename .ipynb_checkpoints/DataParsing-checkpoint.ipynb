Для парсинга был выбран <a href="https://www.100bestbooks.ru"> сайт </a> с топом лучших книг

Импорт библиотек

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

Сохраняем url сайта

```python
url = "https://www.100bestbooks.ru"
```

Далее с помощью библиотеки BeautifulSoup преобразовываем страницу сайта

```python
page = requests.get(url)

text = page.text
soup = BeautifulSoup(text, "lxml")
```

Смотрим на htlm-код сайта

```python
print(soup.prettify())
```

Ищем все куски html-кода с тэгом table

```python
soup.find_all("table")
```

Смотрим на их количество

```python
print(len(soup.find_all("table")))
```

Выбираем только нужную нам таблицу. Для указываем класс table-rating, который только в нужной таблице.

```python
print(len(soup.find_all("table", {"class": "table-rating"})))
```

Далее извлекаем книги из этой таблицы

```python
table = soup.find_all("table", {"class": "table-rating"})[0]

names = []
for t in table.find_all("td", {"class":"vline"}, style="text-align: left"):
    names.append(t.text)
```

Посмотрим на результат

```python
names
```

Далее также делаем для года написания книг

```python
years = []

for t in table.find_all("td", {"class":"vline-year"}):
    years.append(t.text)
```

Убедимся, что количество записей одинаковое

```python
print(len(names))
print(len(years))
```

Теперь объединим данные в data frame

```python
df = pd.DataFrame({'Произведение': names, 'Год': years})

print(df.head(25))
```

Запишем все данные в файл с форматом .csv

```python
df.to_csv("Top_100_books.csv")
```
